## 1_4- Análisis exploratorio inicial de los DataFrames.

Generamos una función con los métodos que deseamos incluir en el EDA Inicial

```python
def resumen_eda(df):
    print("📊 Dimensiones del dataset")
    print(f"Filas: {df.shape[0]}")
    print(f"Columnas: {df.shape[1]}")
    print("-" * 40)

    print("\n🔎 Tipos de datos")
    print(df.dtypes)
    print("-" * 40)

    print("\n🔎 Primeros registros")
    print(df.head(2))
    print("-" * 40)

    print("\n📈 Estadísticas descriptivas")
    print(df.describe(include='number'))
    print("-" * 40)

    print("\n📈 Estadísticas descriptivas")
    try:
        print(df.describe(include=['object', 'str']))
    except:
      print("No hay variables categóricas")
    print("-" * 40)

    print("\n⚠️ Valores nulos por columna")
    nulos = df.isnull().sum()
    if (nulos > 0).any():
        print(nulos[nulos > 0].sort_values(ascending=False))
    else:
        print("No hay valores nulos en el DataFrame (0 en todas las columnas)")
    print("-" * 40)

    print("\n📌 Porcentaje de nulos por columna")
    porcentaje_nulos = df.isnull().mean() * 100
    if (porcentaje_nulos > 0).any():
        print(porcentaje_nulos[porcentaje_nulos > 0].sort_values(ascending=False))
    else:
        print("No hay valores nulos (0% en todas las columnas)")
    print("-" * 40)

    print("\n🔁 Registros duplicados")
    duplicados = df.duplicated().sum()
    print(f"Cantidad de duplicados: {duplicados}")
    print("-" * 40)

```

**Acciones en ventas:**

```python
import pandas as pd

# Crear copia y renombrar columna
df_ven_pp = df_ven.copy()
df_ven_pp.rename(columns={'precio': 'precio_unitario'}, inplace=True)

# Limpieza de precio: quitar '$' y convertir a float
df_ven_pp['precio_unitario'] = df_ven_pp['precio_unitario'].str.replace('$', '', regex=False)
df_ven_pp['precio_unitario'] = pd.to_numeric(df_ven_pp['precio_unitario'], errors='coerce')

# Eliminar duplicados
df_ven_pp.drop_duplicates(inplace=True)

# Eliminar nulos (vimos que hay 2 en precio y 2 en cantidad)
df_ven_pp.dropna(subset=['precio_unitario', 'cantidad'], inplace=True)

# Asegurar que cantidad sea entero
df_ven_pp['cantidad'] = df_ven_pp['cantidad'].astype(int)

print("✅ Limpieza completada.")
display(df_ven_pp.info())
display(df_ven_pp.head())
```

### Análisis Detallado de ROI por Canal
En esta sección calcularemos el ROI para TV, RRSS y Email para identificar cuál es el canal más eficiente.

```python
# 1. Unificar ventas con todas las campañas de marketing
df_roi_full = pd.merge(df_ven_pp, df_mar_pp, on='producto', how='left')

# 2. Identificar ventas que ocurrieron durante la vigencia de la campaña
df_roi_full['venta_en_campanha'] = (
    (df_roi_full['fecha_venta'] >= df_roi_full['fecha_inicio']) &
    (df_roi_full['fecha_venta'] <= df_roi_full['fecha_fin'])
)

# 3. Filtrar solo las ventas atribuidas a campañas y agrupar por canal
roi_por_canal = df_roi_full[df_roi_full['venta_en_campanha']].groupby('canal').agg(
    ingresos_atribuidos=('venta_neta', 'sum'),
    costo_total_campanhas=('costo', 'sum')
)

# 4. Calcular el ROI (Ingresos / Costo)
roi_por_canal['ROI'] = roi_por_canal['ingresos_atribuidos'] / roi_por_canal['costo_total_campanhas']

print("📊 Tabla de Rendimiento por Canal de Marketing:")
display(roi_por_canal.sort_values(by='ROI', ascending=False))
```

```python
# 5. Visualización del ROI por Canal
plt.figure(figsize=(10, 5))
sns.barplot(data=roi_por_canal.reset_index(), x='canal', y='ROI', palette='Set2', hue='canal', legend=False)
plt.axhline(1, color='red', linestyle='--', label='Punto de Equilibrio (ROI=1)')
plt.title('Comparativa de ROI por Canal de Marketing')
plt.ylabel('Retorno de Inversión (x veces)')
plt.xlabel('Canal')
plt.show()
```

### 🔍 Chequeo de Integridad y Consistencia
Realizamos una verificación técnica para asegurar que el preprocesamiento y los análisis posteriores son coherentes.

```python
# 1. Verificación de limpieza
print("--- Verificación de Limpieza ---")
print(f"Nulos en Ventas: {df_ven_pp.isnull().sum().sum()}")
print(f"Duplicados en Ventas: {df_ven_pp.duplicated().sum()}")

# 2. Corrección de Fechas en Marketing
# Identificamos donde fecha_inicio es mayor a fecha_fin
mask = df_mar_pp['fecha_inicio'] > df_mar_pp['fecha_fin']
if mask.any():
    print(f"\n⚠️ Corrigiendo {mask.sum()} campañas con fechas invertidas...")
    # Intercambiamos los valores
    df_mar_pp.loc[mask, ['fecha_inicio', 'fecha_fin']] = df_mar_pp.loc[mask, ['fecha_fin', 'fecha_inicio']].values
    # Recalculamos duración
    df_mar_pp['duracion'] = df_mar_pp['fecha_fin'] - df_mar_pp['fecha_inicio']

# 3. Validación de Merges
ventas_sin_campanha = df_roi_full['canal'].isnull().sum()
print(f"\nVentas no asociadas a ninguna campaña: {ventas_sin_campanha}")

# 4. Consistencia Financiera
total_ventas_base = df_ven_pp['venta_neta'].sum()
total_ventas_roi = roi_por_canal['ingresos_atribuidos'].sum()
print(f"\nTotal Ventas Neta (Base): ${total_ventas_base:,.2f}")
print(f"Total Ventas Atribuidas a Campañas: ${total_ventas_roi:,.2f}")

print("\n✅ ¡Limpieza y corrección de fechas completada!")
```

### 📊 Métricas Finales de ROI (Post-Corrección)
Presentamos los resultados definitivos tras sanear las fechas del dataset de marketing.

```python
# 1. Recalcular la unión y atribución con las fechas corregidas
df_roi_final = pd.merge(df_ven_pp, df_mar_pp, on='producto', how='left')

# Atribución basada en fechas saneadas
df_roi_final['venta_en_campanha'] = (
    (df_roi_final['fecha_venta'] >= df_roi_final['fecha_inicio']) &
    (df_roi_final['fecha_venta'] <= df_roi_final['fecha_fin'])
)

# 2. Agrupación final por canal
roi_final_canal = df_roi_final[df_roi_final['venta_en_campanha']].groupby('canal').agg(
    ingresos_atribuidos=('venta_neta', 'sum'),
    costo_total_marketing=('costo', 'sum'),
    cantidad_ventas=('id_venta', 'count')
)

# 3. Cálculo de ROI final
roi_final_canal['ROI'] = roi_final_canal['ingresos_atribuidos'] / roi_final_canal['costo_total_marketing']

print("📈 Resumen de Performance por Canal (Fechas Corregidas):")
display(roi_final_canal.sort_values(by='ROI', ascending=False).style.format({
    'ingresos_atribuidos': '${:,.2f}',
    'costo_total_marketing': '${:,.2f}',
    'ROI': '{:.2f}x'
}))
```

```python
# 4. Visualización comparativa
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 5))
ax = sns.barplot(data=roi_final_canal.reset_index(), x='canal', y='ROI', palette='viridis', hue='canal', legend=False)

# Añadir etiquetas de valor sobre las barras
for p in ax.patches:
    ax.annotate(f'{p.get_height():.1f}x', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 9), textcoords='offset points', fontsize=11, fontweight='bold')

plt.title('ROI Final por Canal de Marketing', fontsize=14)
plt.ylabel('Retorno (Ventas / Inversión)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
